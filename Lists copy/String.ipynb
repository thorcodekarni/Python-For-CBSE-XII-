<h1>Strings
<h3>Strings are not mutable

```python
s='fc barcelona'

print(s.title())  # Output: Fc Barcelona
print(s.capitalize())  
print(s.upper())
print(s.lower())
print("rEAL mADRID cf".swapcase())
print(s.find('bar'))
print(s.isalnum())
print(s.isalpha())
print(s.isdigit())

```

```python
print(s.replace('barcelona', 'madrid'))
print(s.replace('fc','real'))


#''' only the copies of the string are modified, the original string remains unchanged '''
```

```python
print(s.isspace())
print("  ".isspace())
print(s.startswith('fc'))
print(s.endswith('barcelona'))
print(s.split())
space="  "
print(space.isspace())
print(space.isalnum())
print("abs123".isalnum())

word="Hello, Today is a good day, is not it?."
print(word.partition("is"))
print(word.partition("aka"))
print(type(word.partition("is")))
print(word.split(" "))
print(type(word.split(" ")))
print(word.rpartition("is"))
print(word.rpartition("aka"))
```
