<h1> Traversing through the elements </h1>

```python
trade_logs = [[100,"BTC",110.9,True],200,300]
print(trade_logs[0][1])

#[0] refers to first element of trade_logs
#[1] refers to second element of trade_logs

#[0][1] refers to first element of trade_logs and then 
# second element of that first element

# print(trade_logs[-3][-3])
```

```python
print(trade_logs[-3][-3])

# [-3][-3] does the same as [0][1] because -3 refers to
# third element from last which is the first element in this case
```

```python
# len/min/max/sorted

trade_logs = [[100,"BTC",110.9,True],200,300]
len(trade_logs)

trade_logs = [100,200,300]
min(trade_logs)
max(trade_logs)

sorted(trade_logs, reverse=True)
```

```python

# count 

trade_logs = [100,200,300,100,200,300,100,200,300,100,200,300]
trade_logs.count(100)
```

```python

# reverse
trade_logs = [100,200,300,100,200,300,100,200,300,100,200,300]
trade_logs.reverse()
trade_logs

# permanently reverses the list
```

```python
# sort (vs sorted)
trade_logs = [100,200,300,100,200,300,100,200,300,100,200,300]
sorted(trade_logs)
print(trade_logs)
trade_logs.sort()
print(trade_logs)
```

```python

# copy -> shallow
trade_logs = [100,200,300,100,200,300,100,200,300,100,200,300]
print(trade_logs)
print(id(trade_logs))
new_trade_logs = trade_logs.copy()
print(new_trade_logs)
print(id(new_trade_logs))
```
