<h1> Traversing through the elements </h1>

```python
trade_logs = [[100,"BTC",110.9,True],200,300]
print(trade_logs[0][1])

#[0] refers to first element of trade_logs
#[1] refers to second element of trade_logs

#[0][1] refers to first element of trade_logs and then 
# second element of that first element

# print(trade_logs[-3][-3])
```

```python
print(trade_logs[-3][-3])

# [-3][-3] does the same as [0][1] because -3 refers to
# third element from last which is the first element in this case
```

<h3>SLICING

```python
trade_logs = [[100,"BTC",110.9,True],200,300]
print(trade_logs[0][::2])  #slicing to get alternate elements from first element of trade_logs


```

```python
print(trade_logs[-3:-1])  #slicing to get first two elements of trade_logs])    
```
