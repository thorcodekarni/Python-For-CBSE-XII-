```python
c=complex(input("Enter a complex number in the form a+bj: "))
print(c)
print("The complex number is:", c)

print("The real part is:", c.real)
print("The imaginary part is:", c.imag) 

print("The conjugate is:", c.conjugate())
d= c.conjugate()
print("The conjugate of conjugate is:", d.conjugate())
```
