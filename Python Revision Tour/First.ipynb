```python
a= 4
b= 7
print(a&b)  # Bitwise AND
print(a|b)  # Bitwise OR
print(a^b)  # Bitwise XOR
print(a>b)  # Comparison Operators
print(a<b)  # Comparison Operators
print(a==b)
print(a!=b)
```

```python
print(type(a>b))    # To check the data type of the result of comparison operators  
print(type(a<b))    # To check the data type of the result of comparison operators
print(type(a==b))
print(type(a!=b))
print(type(a&b))    #binary operators result data type
```
